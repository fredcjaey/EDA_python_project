```python
#Task 1

years = ["2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"]
durations = [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]
movie_dict = dict(zip(years, durations))
print(movie_dict)
```

```python
#Task 2

import pandas as pd
durations_df = pd.DataFrame(list(movie_dict.items()))
durations_df.columns = ['years', 'duration']
print(durations_df)
products_csv = durations_df.to_csv('products.csv', index=True)
```

```python
#Task 3

import matplotlib.pyplot as plt
plt.plot(years, durations)
plt.xlabel ('Years')
plt.ylabel ('Duration')
plt.title ('Netflix Movie Durations 2011-2020')
```

```python
#Task 4

netflix_df = pd.read_csv('netflix_data.csv')
print(netflix_df.head(5))
```

```python
#Task 5

netflix_df_movies_only = netflix_df[netflix_df["type"] == "Movie"]
netflix_movies_col_subset = netflix_df_movies_only[["title", "country", "genre", "release_year", "duration"]]
print(netflix_movies_col_subset.head(5))
```

```python
#Task 6

plt.scatter(netflix_movies_col_subset["release_year"], netflix_movies_col_subset["duration"])
plt.xlabel ('Years')
plt.ylabel ('Duration')
plt.title ('Movie Duration by Year of Release')
plt.show()
```

```python
#Task 7

short_movies = netflix_movies_col_subset[netflix_movies_col_subset["duration"] < 60]
print(short_movies.head(20))
```

```python
#Task 8

colors = []
for genre in netflix_movies_col_subset["genre"]:
    if genre == "Children":
        colors.append("red")
    elif genre == "Documentaries":
        colors.append("blue")
    elif genre == "Stand-Up":
        colors.append("green")
    else:
        colors.append("black")
print(colors[:10])
```

```python
#Task 9

fig = plt.figure(figsize=(16,12))
plt.style.use('fivethirtyeight')
plt.scatter(netflix_movies_col_subset["release_year"], netflix_movies_col_subset["duration"],color=colors)
plt.title('Movie duration vs year of release by genre')
plt.xlabel('Release year')
plt.ylabel('Duration (min)')
plt.show()
```
